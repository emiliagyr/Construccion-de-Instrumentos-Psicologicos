#3. Correlación

```python
pg.pairwise_corr(data=df, columns=["CLIMA", "CARGA_DE_TRABAJO"])
```

**Coeficiente de Correlación (r):**

El coeficiente de correlación (r) varía de -1 a 1.

Un valor positivo de r indica una correlación positiva, lo que significa que a medida que una variable aumenta, la otra tiende a aumentar también.

Un valor negativo de r indica una correlación negativa, lo que significa que a medida que una variable aumenta, la otra tiende a disminuir.

Cuanto más cercano sea el valor de r a 1 o -1, más fuerte será la correlación. Un valor de 0 indica ausencia de correlación lineal.

**Significancia (valor p):**

El valor p asociado con cada coeficiente de correlación indica la probabilidad de observar tal correlación si no hubiera una verdadera asociación entre las variables.

Típicamente, si el valor p es menor que un nivel de significancia elegido (por ejemplo, 0.05), se considera que la correlación es estadísticamente significativa.

Si el valor p es mayor que el nivel de significancia, la correlación no se considera estadísticamente significativa.

**Fuerza de la Correlación:**

La magnitud del coeficiente de correlación indica la fuerza de la relación entre las variables.
Las pautas comunes para interpretar la fuerza de los coeficientes de correlación son:

0.00-0.19: Correlación muy débil

0.20-0.39: Correlación débil

0.40-0.59: Correlación moderada

0.60-0.79: Correlación fuerte

0.80-1.00: Correlación muy fuerte
